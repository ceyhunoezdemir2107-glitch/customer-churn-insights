# Customer Churn EDA

Goal: understand churn patterns, identify business-relevant drivers, and prepare modeling decisions.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

DATA_PATH = Path('../data/raw/telco_customer_churn.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
```

```python
df['Churn'].value_counts(normalize=True).rename('share')
```

```python
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```
